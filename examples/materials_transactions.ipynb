# Cymetric Example for materials exchanges

This example walks through how to extract and plot information about materials exchanges in a Cyclus simulation

### Learning Objectives:
* Open an SQLite database with Cymetric
* Store a metric as a pandas DataFrame
* Use methods available in cymetric
* Plot the data in a DataFrame  

```python
import cymetric as cym
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from cymetric import filters as cyf
from cymetric import graphs as cyg
from cymetric import timeseries as cyt
%matplotlib inline
```

```python
#Remove any current database with the same name, run cyclus to have the most updated output database
! rm recycle.sqlite
! cyclus recycle.xml -o recycle.sqlite #-o flag specifies output file name
```

```python
db = cym.tools.dbopen('recycle.sqlite')
evaler = cym.Evaluator(db, write=False)
transactions = evaler.eval('TransactionQuantity').sort_values(by='TimeCreated')
transactions = transactions.rename(columns={'TimeCreated':'Time'})
```

### Using Cymetric we can look at how materials flow from one facility to another 

```python
cyg.flow_graph(evaler, label='com')
```

This flow graph shows how the different commodities flow between each facility type. Notice that the repository only receives the waste commodity, while all spent fuel goes to the lwr_separations facility. 

### We can plot the amount of material sent to the repository as a function of time

Let's use the filters.transactions method to look at each of the material transactions in which the repository is the receiving facility.   

```python
repo_waste = cyt.transactions(evaler=evaler, receivers=['repo'])
```

```python
repo_waste.plot(x='Time', figsize=(20,10), legend=False)
plt.xlabel('Time (months)',fontsize=18)
plt.ylabel('Mass of waste (kg)',fontsize=18)
plt.savefig('waste.png')
```

```python
repo_waste.plot(x='Time', figsize=(20,10), legend=False, logy=True)
plt.xlabel('Time (months)',fontsize=18)
plt.ylabel('Mass of waste (kg)',fontsize=18)
```

These figures shows the amount of waste sent to the repository at each time step. It is clear that there is some waste being sent to the repository at each time step. Looking at the graph with a logarithmic scale, there is a large decrease of waste sent to the repository once teh LWRs are decommissioned, due to waste from the FRX Separations facilities. 

If we want to look at the cumulative amount of waste sent to the repository, we need to first take a cumulative sum of the waste quantity. Then we can create the plot in a similar manner. 

```python
repo_waste['Mass'] = repo_waste['Mass'].cumsum()
repo_waste.plot(x='Time', figsize=(20,10))
plt.xlabel('Time (months)',fontsize=18)
plt.ylabel('Mass of waste (kg)',fontsize=18)
plt.savefig('waste_cumulative.png')
```

### We can look at the amount of fuel used in the simulation
Let's look at the cumulative amount of fuel fabricated in the simulation. To do this, we need to find the transactions that involve each fuel type (uox or frx_fuel), and take a cumulative sum. Since not all transactions will involve a fuel type a NaN value will be placed in the rows in which the specified fuel type is not the commodity being traded. Therefore, the NaN values must be filled in with the cumulative amount from the last transaction of the desired commodity.

```python
transactions['uox_total'] = transactions.loc[transactions['Commodity'] == 'uox']['Quantity'].cumsum()
transactions['frx_total'] = transactions.loc[transactions['Commodity'] == 'frx_fuel']['Quantity'].cumsum()
transactions.fillna(method='ffill', inplace=True)
transactions['total_fuel'] = transactions['uox_total'] + transactions['frx_total']
```

```python
transactions[['Time', 'uox_total', 'frx_total', 'total_fuel']].plot(x = 'Time', figsize=(20,10), legend=False)
plt.xlabel('Time (months)',fontsize=18)
plt.ylabel('Mass of fuel fabricated (kg)',fontsize=18)

l = plt.legend()
l.get_texts()[0].set_text('UOX')
l.get_texts()[1].set_text('FRX')
l.get_texts()[2].set_text('Total')
l.set_title('Fuel Type')

plt.savefig('fuel_cumulative.png')
plt.show()
```

Using the cymetric transactions tool, we can also look at specific nuclides in each transaction. Let's look at the cumulative U-235 that is received by all reactor facilities 

```python
U235 = cyt.transactions(evaler=evaler, receivers=['lwr', 'frx'], nucs=['92235'])
U235['Mass'] = U235['Mass'].cumsum()
U235.plot(x='Time', figsize=(20,10), legend=False, )
plt.xlabel('Time (months)',fontsize=18)
plt.ylabel('Mass of U235 (kg)',fontsize=18)
plt.savefig('U235_mass.png')
```
