# Cymetric Example for Build and Decommissioning of Prototypes

This example walks through how to plot the number of reactor prototypes built during the simulation. 
The exercise starts by running the given input file, 'recycle.xml' through Cyclus to give an output database 'recycle.sqlite'. 

### Learning Objectives:
* Open a SQLite Database with Cymetric
* Store a metric as a pandas DataFrame
* Manipulate the contents and column names of a DataFrame
* Plot the data in a DataFrame in a bar chart

```python
import cymetric as cym
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Remove any current database with the same name, run cyclus to have the most updated output database
! rm recycle.sqlite
! cyclus recycle.xml -o recycle.sqlite #-o flag specifies output file name
```

```python
db = cym.dbopen('recycle.sqlite')
evaler = cym.Evaluator(db)
comm = evaler.eval('BuildSeries')
decomm = evaler.eval('DecommissionSeries') 
```

A full list of metrics avilable through Cymetric can be found in [cymetric/metrics.py](https://github.com/cyclus/cymetric/blob/master/cymetric/metrics.py) in the cymetric repository. General metrics begin on line 94 of the code. 

### We can plot the number of reactor prototypes entering and exiting the simulation at each time step

To do this we change the decommissiong values to negative, rename the time step columns to match in each DataFrame, reorient the tables based on the prototype and merge the build and decommission data. Finally, the data is plotted as a function of the time step. 

```python
neg = -decomm['Count']
decomm = decomm.drop('Count',axis=1)
decomm = pd.concat([decomm, neg],axis=1)
```

```python
comm = comm.rename(columns={'EnterTime':'Time'})
decomm = decomm.rename(columns={'ExitTime':'Time'})
```

```python
c = comm.pivot('Time','Prototype')['Count'].reset_index()
d = decomm.pivot('Time','Prototype')['Count'].reset_index()
cd = pd.merge(c,d,left_on='Time',right_on='Time',how='outer',sort=True, \
              suffixes=('_enter','_exit')).fillna(0)
```

```python
#pandas always changes everything to floats in the presence of NaN, so changing TimeStep values to integers
cd.Time = cd.Time.astype(int)
```

```python
#Plotting facilities of interst: lwr and frx
colors = ['green','purple','orange']
ax = cd[['Time','lwr_enter','frx_enter']].plot.bar(x='Time',figsize=(20,10),stacked=True, \
                                               legend=False, color=colors)
l = plt.legend()
l.get_texts()[0].set_text('LWR')
l.get_texts()[1].set_text('FR')
l.set_title('Prototype')
cd[['Time', 'lwr_exit','frx_exit']].plot.bar(x='Time', ax=ax, figsize=(20,10),stacked=True, \
                                               legend=False, color=colors)
plt.xlabel(r'TimeStep',fontsize=18); plt.ylabel(r'Count',fontsize=18)
plt.show()
#plt.savefig('ex.png')
```

Note that in this plot, only the time steps in which a prototype is built or decommissioned is included. 

### Alternatively, we can plot the number of reactor prototypes as a function of the time step

To do this, we will first have to sum the number of each reactor prototype entering and exiting the simulation at each time step. Then we will sum the cumulative totals of the prototypes entering and exiting the simulation. 

```python
# get cumulative sums of _enter and_exit columns for reactors, then adding together
for rx in ['lwr','frx']:
    cd[str(rx) + '_total'] = cd[str(rx)+'_enter'].cumsum() + cd[str(rx)+'_exit'].cumsum()
```

```python
#Plotting facilities of interst: lwr and frx
colors = ['green','purple']
ax = cd[['Time','lwr_total','frx_total']].plot(x='Time',figsize=(20,10), \
                                               legend=False, color=colors)
l = plt.legend()
l.get_texts()[0].set_text('LWR')
l.get_texts()[1].set_text('FR')
l.set_title('Prototype')
plt.xlabel(r'TimeStep',fontsize=18); plt.ylabel(r'Count',fontsize=18)
plt.show()
#plt.savefig('cumulative_ex.png')
```
